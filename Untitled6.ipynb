```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.cm as cm
```

```python
x=np.array([1, 2, 3,4,5,6,7])
y=np.array([1, 2, 3,5,8,13,20])
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x, marker="o", markersize=10, linestyle= 'dashed', linewidth=3, label='Normal' )
plt.plot(y, marker="o", markersize=10, linestyle= 'dashed',linewidth=3, label='Fast' )
plt.legend()
```
